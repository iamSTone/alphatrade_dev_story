1. 가격 정보 가져오기

* 난이도가 높은 순서대로: 증권사 API > KOSCOM OpenAPI > Yahoo API

### 증권사 API
* 대신증권, 이트레이드투자증권, 키움증권 등 증권사에서 제공하는 API를 사용하여 가격 정보를 얻어오는 방식입니다.
* 반드시 MS 윈도우를 사용해야 하며, 증권사 계좌 및 공인인증서가 필요합니다.
    - 키움증권은 2개월 단위로 천만원 ~ 일억원이 들어있는 가상계좌를 이용한 모의투자를 제공합니다.
* C++로 만들어진 COM 기반 컴포넌트를 제공하는데, 이를 Python으로 호출하는 방식이 비교적 용이합니다.
    - 이에 대한 자세한 방법은 [이곳](https://wikidocs.net/book/110)을 참고하시기 바랍니다.

### KOSCOM OpenAPI

#### 서비스 사용 등록 및 키 발급
* [코스콤](https://fintech.koscom.co.kr/portal/main/contents.do?menuNo=200050)에서 REST 기반 웹 API를 베타테스트 중입니다. (7월 28일 종료, 8월 중 정식 오픈 예정)
* [사용신청서](https://fintech.koscom.co.kr/portal/cmm/fms/FileDown.do?atchFileId=FILE_000000000001635&fileSn=1&bbsId=)를 이메일로 제출하면 [플랫폼](https://testbed.koscom.co.kr/portal)에 아이디를 부여해 줍니다.
* 로그인 후 [내 앱 관리](https://testbed.koscom.co.kr/portal/api-manager/users/alphatrade/apps) 메뉴에서 앱을 생성한 다음 구독할 서비스를 설정하면 구독 서비스에 따라 API 사용을 위한 키를 발급받아 사용합니다.
* 서비스 종류
    - stocks (주식): 주식시장에서 발생하는 체결가, 호가 등의 거래정보와 회원사별매매동향 등의 정보를 실시간으로 제공
    - futures (선물): 주가지수 상품에서 발생하는 체결가, 호가 등의 실시간 거래정보와 종목의 기초자산 및 정산이론 가격등의 상세정보 제공합니다.
    - options (옵션): 주가지수 상품에서 발생하는 체결가, 호가 등의 실시간 거래정보와 종목의 기초자산 및 정산이론 가격등의 상세정보 제공합니다.
    - industry (업종): 코스피/코스닥 업종지수 데이터 제공합니다.
    - news (시황): 다양한 금융 상품이나 주제,키워드에 대한 뉴스들의 정보를 제공합니다. 투자자는 본인의 관심사에 따라 뉴스를 검색하여 찾거나, 지정한 키워드에 따른 뉴스를 개인화 하여 제공 받을 수 있습니다. 현재 1990년부터 25년치의 뉴스 약 7천만건을 제공합니다. (위버플 제공)

#### 예제 코드
* 현재가 받아오기

````python
import json
import urllib
import urllib2

import secrete    # API 키가 저장되어 있는 모듈 (비밀)

_PRICE_URL = 'https://testbed.koscom.co.kr/gateway/v1/market/stocks/price'    # 가격 정보 접근 URL

stock = '005930'    # 삼성전자
# URL에 API키 및 종목 파라미터를 세팅
url = '%s?%s' % (_PRICE_URL, urllib.urlencode({'apikey': secrete.STOCKS_API_KEY,    # API 키
                                               'isuSrtCd': stock}))                 # 종목 코드
json_obj = json.load(urllib2.urlopen(url))    # JSON 파싱
print 'price of %s is %d' % (stock, json_obj['trdPrc'])    # 가격 정보 출력
````

* 종가 받아오기

````python
import json
import urllib
import urllib2

import secrete    # API 키가 저장되어 있는 모듈 (비밀)

_HISTORY_URL = 'https://testbed.koscom.co.kr/gateway/v1/market/stocks/history'    # 종가 정보 접근 URL

stock = '005930'    # 삼성전자
# URL에 API키 및 종목 파라미터를 세팅
url = '%s?%s' % (_HISTORY_URL, urllib.urlencode({'apikey': secrete.STOCKS_API_KEY,    # API 키
                                                 'isuSrtCd': stock,                   # 종목 코드
                                                 'trnsmCycleTpCd': 'D',               # 주기 (D:일별, W:주별, M:월별)
                                                 'inqStrtDd': '20160701',             # 시작 날짜
                                                 'inqEndDd': '20160715',              # 종료 날짜
                                                 'reqCnt': 100}))                     # 요청 최대 갯수
json_obj = json.load(urllib2.urlopen(url))    # JSON 파싱
for day_obj in json_obj['inqLists']:
    # 날짜별 가격 정보 출력
    print 'date: %s, close price: %d' % (day_obj['trdDd'], day_obj['clsprc'])
````

    - 종가는 가장 최신 날짜를 기준으로 한 수정 종가입니다. (주의!)
